```python
df['born_date']= df.Born.apply(lambda x :''.join(x.strip().split()[:3]) if isinstance(x,str) else None )
df['death_date']= df.Died.apply(lambda x : ''.join(x.strip().split()[:3]) if isinstance(x,str) else None)
df['born_date']= pd.to_datetime(df.born_date, format= '%d%B%Y' ,errors='coerce')
df['death_date']= pd.to_datetime(df.death_date, format= '%d%B%Y' ,errors='coerce')
```

```python
df.Born.str.strip().str.split()[:3]
```

```python
df['heights(cm)']= df.Measurements.apply(lambda x : x.strip().split()[0] if isinstance(x,str) else None)
df['weights(kg)']= df.Measurements.apply(lambda x : x.strip().split()[3] if isinstance(x,str) and len(x.strip().split())>4 else None)
df["heights(cm)"] = pd.to_numeric(df["heights(cm)"], errors="coerce")
df["weights(kg)"] = pd.to_numeric(df["weights(kg)"], errors="coerce")
```

```python
pd.set_option('display.max_colwidth', None)
print(df['Born'].iloc[0])
```

```python
import numpy as np
df['r']= df.Born.apply(lambda x : x.split('in')[1] if isinstance(x,str) and len(x.strip().split('in'))>1 else None)
df['country']=df.r.str.extract(r"\(([^()]*)\)\s*$")
df["loc_clean"] = df.r.str.replace(r"\s*\([^()]*\)\s*$", "", regex=True)

df[["city", "region"]] = df["loc_clean"].str.split(",", n=1, expand=True)
df["city"] = df["city"].str.strip().replace("?",np.nan)
df["region"] = df["region"].str.strip()
```

```python
df.head(2)
```

```python
df['name']= df['Used name'].str.replace('•',' ')
```

```python
df.head(1)
```

```python
col_to_keep= ['athlete_id', 'name', 'Sex', 'born_date','death_date',
              'NOC', 'heights(cm)', 'weights(kg)','region', 'city' ,'country']
df_clean= df[col_to_keep]
df_clean.head()
```
